```python
import os
from functools import reduce
import chardet

def list_docs(folder_path):
    """Return a list of strings, where each string is the contents of one document in the folder."""
    docs = []
    for filename in os.listdir(folder_path):
        with open(os.path.join(folder_path, filename), "rb") as f:
            data = f.read()
            encoding = chardet.detect(data)['encoding']
        with open(os.path.join(folder_path, filename), "r", encoding=encoding) as f:
            docs.append(f.read())
    return docs

def index_documents(folder_path):
    # Map phase
    word_counts = []
    for doc_id, doc in enumerate(list_docs(folder_path)):
        doc_word_counts = {}
        for word in doc.split():
            if word in doc_word_counts:
                doc_word_counts[word] += 1
            else:
                doc_word_counts[word] = 1
        word_counts.append((doc_id, doc_word_counts))

    # Reduce phase
    word_index = {}
    for doc_id, doc_word_counts in word_counts:
        for word, count in doc_word_counts.items():
            if word in word_index:
                word_index[word].append((doc_id, count))
            else:
                word_index[word] = [(doc_id, count)]

    return word_index

def search_word(word_index, word):
    if len(word.split()) > 1:
        print("Error: Please enter only one word.")
        return
    if word not in word_index:
        return f"The word '{word}' is not found in any document."
    word_occurrences = word_index[word]
    total_count = reduce(lambda x, y: x + y, (count for _, count in word_occurrences))
    result = f"The word '{word}' is repeated {total_count} times"
    for doc_id, count in word_occurrences:
        result += f"\n{count} times in doc {doc_id + 1}"
    return result

word_index = index_documents('/content/drive/My Drive/Assignment2data')
word = input("Enter the word to look for: ")
print(search_word(word_index, word))
```
